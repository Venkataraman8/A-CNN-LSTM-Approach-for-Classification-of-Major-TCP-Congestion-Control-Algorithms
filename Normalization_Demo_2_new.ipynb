```
new = df_x.values.tolist() ## Dataframe to list.....

List1 = []
for x in range(len(new)):
    temp = []
    temp = new[x]
    if 'A' in temp:
        temp1 = []
        for m in temp:
            if m != 'A':
                temp1.append(m)
        List1.append(temp1)        
    if 'A' not in temp:
        List1.append(temp)

List1
```

```
for i in range(len(List1)):
  for j in range(len(List1[i])):
    List1[i][j]=math.trunc(List1[i][j])
 
```

```
print(List1)
```

```
for i in range(len(List1)):
    Wmin.append(min(List1[i]))
    Wmax.append(max(List1[i]))
```

```
print(Wmin)
```

```
#Basheer Algorithm for Min-Max Normalization, Lambda1 = 0.1, Lamda2 = 0.9
```

```
for i in range(len(List1)):
    for j in range(len(List1[i])):
       Wmin[i] = abs(Wmin[i])
       Wmax[i] = abs(Wmax[i])
       a = (0.8*(List1[i][j] - Wmin[i]))
       b= (Wmax[i] - Wmin[i])
       c= a/b
       List1[i][j]= 0.1+ c
```

```
for i in range(len(List1)):
    for j in range(len(List1[i])):
       Wmin[i] = abs(Wmin[i])
       Wmax[i] = abs(Wmax[i])
       a = ((List1[i][j] - Wmin[i]))
       b= (Wmax[i] - Wmin[i])
       c= a/b
       List1[i][j]= c
```

```
rows, cols = (len(List1), 128) 
I = [[0]*cols]*rows 
for i in range(len(List1)):
          temp=[]
          for u in range(128):
                      surj = 128/len(List1[i])
                      vall = u/surj
                      ind = math.trunc(vall)
                      temp.append(List1[i][ind])
          I[i]=temp
```
